## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset

df = pd.merge(mouse_metadata, study_results, how="inner", on="Mouse ID")

# Display the data table for preview

df.head(10)
```

```python
df = df.rename(columns={"Mouse ID":"mouse", "Drug Regimen": "drug",
                       "Sex":"sex", "Age_months":"age", "Weight (g)":"weight",
                       "Timepoint":"timepoint", "Tumor Volume (mm3)":"tum_vol", 
                       "Metastatic Sites": "m_sites"})
df
```

```python
# Checking the number of mice.
df["mouse"].nunique()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

df[df.duplicated(subset=["mouse", "timepoint"])]
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

df[df["mouse"] == "g989"]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

clean_df = df[df["mouse"] != "g989"]
```

```python
# Checking the number of mice in the clean DataFrame.
clean_df["mouse"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

drugs = clean_df["drug"].unique()
print(sorted(drugs))


df_dict = {}


for drug in drugs:
    df = clean_df[clean_df["drug"] == drug]
    df_dict[drug] = df.sort_values(by="mouse")
```

```python
df_dict["Capomulin"].reset_index(drop=True)
df_dict["Ramicane"].reset_index(drop=True)
df_dict["Infubinol"].reset_index(drop=True)
df_dict["Ceftamin"].reset_index(drop=True)
```

```python
capo_s = df_dict["Capomulin"]["tum_vol"].agg(["mean", "median", "var", "std", "sem"])
ceft_s = df_dict["Ceftamin"]["tum_vol"].agg(["mean", "median", "var", "std", "sem"])
infu_s = df_dict["Infubinol"]["tum_vol"].agg(["mean", "median", "var", "std", "sem"])
keta_s = df_dict["Ketapril"]["tum_vol"].agg(["mean", "median", "var", "std", "sem"])
naft_s = df_dict["Naftisol"]["tum_vol"].agg(["mean", "median", "var", "std", "sem"])
plac_s = df_dict["Placebo"]["tum_vol"].agg(["mean", "median", "var", "std", "sem"])
prop_s = df_dict["Propriva"]["tum_vol"].agg(["mean", "median", "var", "std", "sem"])
rami_s = df_dict["Ramicane"]["tum_vol"].agg(["mean", "median", "var", "std", "sem"])
stel_s = df_dict["Stelasyn"]["tum_vol"].agg(["mean", "median", "var", "std", "sem"])
zoni_s = df_dict["Zoniferol"]["tum_vol"].agg(["mean", "median", "var", "std", "sem"])

summary = pd.DataFrame(
    data=[capo_s, ceft_s, infu_s, keta_s, naft_s, plac_s, prop_s, rami_s, stel_s, zoni_s], 
    index=sorted(drugs))

summary
```

```python
for drug in sorted(drugs):
    
    mean = df_dict[drug]["tum_vol"].mean()
    median = df_dict[drug]["tum_vol"].median()
    var = df_dict[drug]["tum_vol"].var()
    std = df_dict[drug]["tum_vol"].std()
    sem = df_dict[drug]["tum_vol"].sem()
    data = [mean, median, var, std, sem]
    print(data)
```
